```python
df['State or Province'].value_counts()[:5]
```

```python
top_states = df.groupby("State or Province")["Sales"].sum().sort_values(ascending=False).head(5).reset_index()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="State or Province", y="Sales", data=top_states)
plt.title("Top 5 States by Sales")
plt.show()
```
